### обработка файла

```python
COLUMNS = ["x1", "Реестровый номер контракта", "x2", "x3", "x4", "x5",
                      "x6", "x7", "x8", "x9", "x10", "x11", "x12",
                      "date1", "date2", "date3", "date4", "date5", "date6", "x13", "x14",
                      "x15", "x16", "x17", "x18", "y_description", "y"]
# пишем результат на диск
PATH = r"C:\Users\AnnaZheleznova\Documents\Cource_BigData\Diplom_github\Data" 
SEP = "___"
```

```python
# датасет объемный, используем параметр chinksize, делим на строки
chunksize = 10 ** 6
i = 0
with pd.read_csv('C:\\Users\\AnnaZheleznova\\Documents\\Cource_BigData\\Diplom_github\\fz.zip', chunksize=chunksize, on_bad_lines="skip",dtype='unicode') as reader:
    for df in reader:
        if i < 3:
            i += 1
            continue

        df.columns = COLUMNS

        df["y"] = df["y"].astype(str)
        df = df[df["y"] != "<НЕ ОПРЕДЕЛЕНО>"]
        df = df[df["y"] != "<НЕ РАСПРЕДЕЛЕНО>"]
        df = df[df["y"].str.len() == 4]

        break
```

```python
# обрабатываем и сохраняем данные в PATH, немного ждем результат
chunksize = 10 ** 5
chunk_id = 0
with pd.read_csv('C:\\Users\\AnnaZheleznova\\Documents\\Cource_BigData\\Diplom_github\\fz.zip', chunksize=chunksize, on_bad_lines="skip", dtype='unicode') as reader:
    for df in reader:
        print(chunk_id, end=", ")
        if chunk_id >= 0:
            df.columns = COLUMNS

            df["y"] = df["y"].astype(str)
            df = df[df["y"] != "<НЕ ОПРЕДЕЛЕНО>"]
            df = df[df["y"] != "<НЕ РАСПРЕДЕЛЕНО>"]
            df = df[df["y"].str.len() == 4]


            df = df[["x4", "x7", "x8", "x11", "x12", "x14", "x15", "y_description", "y"]]

            for y in df["y"].unique():
                if len(y) < 20:
                    df_crop = df[df["y"] == y]
                    df_crop.to_csv(os.path.join(PATH, f"{y}{SEP}{chunk_id}.csv"))
                else:
                    print()
                    print(y)

        chunk_id += 1
```
